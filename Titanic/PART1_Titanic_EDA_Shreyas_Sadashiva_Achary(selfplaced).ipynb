## Survivors interms of gender

```python
fig,axes=plt.subplots(nrows=1, ncols=2,figsize=(10, 4))
women = eda[eda['Sex']=='female']
men = eda[eda['Sex']=='male']
ax = sns.distplot(women[women['Survived']==1].Age.dropna(), bins=20, label = 'survived',ax=axes[0], kde =False)
ax = sns.distplot(women[women['Survived']==0].Age.dropna(), bins=45, label = 'dead',ax=axes[0], kde =False)
ax.legend()
ax.set_title('Female')
ax = sns.distplot(men[men['Survived']==1].Age.dropna(), bins=20, label = 'survived',ax=axes[1], kde =False)
ax = sns.distplot(men[men['Survived']==0].Age.dropna(), bins=45, label = 'dead',ax=axes[1], kde =False)
ax.legend()
ax.set_title('male')
```

```python
eda.groupby('Sex')[['Survived']].mean()
```

From this we conclude that there were more female survivors compared to male survivors
